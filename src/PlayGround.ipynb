```python
def load_image(img_dir='../imgs/'):
    """
    Load image and label data
    @ param img_dir: directory pass for image folder
    @ return img_l: numpay array of image
    @ return lbl_l: numpay array of label
    """
    img_names = os.listdir(img_dir)[0:3]
    img_l = []
    lbl_l = []
    for img_name in img_names:
        lbl_l.append(int(img_name[0]))
        print(img_name[0])
        img = cv2.imread(str(img_dir + img_name))
        b,g,r = cv2.split(img)
        img = cv2.merge([r,g,b])
        img_l.append(img)
    
    img_l = np.array(img_l)
    lbl_l = np.array(lbl_l)
    return img_l, lbl_l
```

```python
img_l, lbl_l = load_image()
```
